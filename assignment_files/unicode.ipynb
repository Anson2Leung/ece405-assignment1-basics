Problem (unicode2): Unicode Encodings (3 points)

```python
ts1 = "hello! こんにちは!"
ts2 = "hello! 你好!"
ts3 = "abc ;.,"
test = [ts1, ts2, ts3]
enc_8 = [s.encode("utf-8") for s in test]
enc_16 = [s.encode("utf-16") for s in test]
enc_32 = [s.encode("utf-32") for s in test]
```

```python
for i, s in enumerate(test):
    print(f"--- '{s}' ---")
    print(len(enc_8[i]), enc_8[i])
    print(len(enc_16[i]), enc_16[i])
    print(len(enc_32[i]), enc_32[i])
```

```python
def decode_utf8_bytes_to_str_wrong(bytestring: bytes):
        return "".join([bytes([b]).decode("utf-8") for b in bytestring])
```

```python
print(decode_utf8_bytes_to_str_wrong("hello".encode("utf-8")))
print("hello! 你好!".encode("utf-8"))
print("hello! 你好!".encode("utf-8").decode("utf-8"))
print(decode_utf8_bytes_to_str_wrong("hello! 你好!".encode("utf-8")))
```
